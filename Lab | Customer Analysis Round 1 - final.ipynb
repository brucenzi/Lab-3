```python
#10 Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
```

```python
dfconcat
print(dfconcat.dtypes)
```

```python
#as the symbol "%" is not read as numerical character, it must be removed before making a numerical operation.
#applying the "replace" function to remove the %" in the whole column will convert the existent numerical data in "nan", then it is needed to convert the whole column to string first
dfconcat["customer lifetime value"] = dfconcat["customer lifetime value"].apply(str)
dfconcat
```

```python
#now that the column is converted to str, remove the "%" symbol and the "nan" data
dfconcat["customer lifetime value"] = dfconcat["customer lifetime value"].str.replace("%","")
dfconcat
```

```python
#now that the column is converted to str, remove the "%" symbol and the "nan" data
dfconcat["customer lifetime value"] = dfconcat["customer lifetime value"].str.replace("nan","")
dfconcat
```

```python
#now the column is clean to be converted to numerical data. use pd.to_numeric to convert to integer or float when it applies.
dfconcat["customer lifetime value"] = pd.to_numeric(dfconcat["customer lifetime value"])
dfconcat
```

```python
#function to multiply the values of a column
dfconcat["customer lifetime value"] = 100 * dfconcat["customer lifetime value"]
dfconcat
```
